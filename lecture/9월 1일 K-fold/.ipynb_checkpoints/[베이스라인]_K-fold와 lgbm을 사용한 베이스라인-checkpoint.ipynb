```python
from sklearn.cluster import k_means
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

d_list = []
for i in range(1, 61):
    d = train[train['num'] == i].copy()
    d['전력사용량(kWh)'] = scaler.fit_transform(d['전력사용량(kWh)'].values.reshape(-1,1))
    d_ = d.groupby(['weekday', 'hour'])['전력사용량(kWh)'].mean().unstack()
    d_list.append(d_.to_numpy())

# 정규화된 시간-요일별 전력사용량을 이미지처럼 저장
d_list = np.array(d_list)
```

```python
# 불쾌지수 컬럼을 train.test 에 새컬럼으로 만들어보세요.
train["THI"]=1.8*train['temp']-0.55*(1 - 0.01*train['humidity'])*(1.8*train['temp']-26)+32
test["THI"]=1.8*test['temp']-0.55*(1 - 0.01*test['humidity'])*(1.8*test['temp']-26)+32
```

```python
# THI one-hot 
# THI_1 : THI < 68
# THI_2 : 68 <= THI < 75
# THI_3 : 75 <= THI < 80
# THI_4 : 80 <= THO
```

```python
train['THI_1']=(train['THI']<68).astype(int)
train['THI_2']=((train['THI']>=68)&(train['THI']<75)).astype(int)
train['THI_3']=((train['THI']>=75)&(train['THI']<80)).astype(int)
train['THI_4']=(train['THI']>=80).astype(int)

test['THI_1']=(test['THI']<68).astype(int)
test['THI_2']=((test['THI']>=68)&(test['THI']<75)).astype(int)
test['THI_3']=((test['THI']>=75)&(test['THI']<80)).astype(int)
test['THI_4']=(test['THI']>=80).astype(int)
```

```python
# THI_Type 칼럼을 만들어 봅시다. 불쾌지수가 1단계 -> 1 , 2단계 -> 2 , 3단계 -> 3 , 4단계 -> 4 인 컬럼

train.loc[train['THI']<68,'THI_type'] = 1
train.loc[(train['THI']>=68)&(train['THI']<75),'THI_type'] = 2
train.loc[(train['THI']>=75)&(train['THI']<80),'THI_type'] = 3
train.loc[train['THI']>=80,'THI_type'] = 4

test.loc[test['THI']<68,'THI_type'] = 1
test.loc[(test['THI']>=68)&(test['THI']<75),'THI_type'] = 2
test.loc[(test['THI']>=75)&(test['THI']<80),'THI_type'] = 3
test.loc[test['THI']>=80,'THI_type'] = 4
```

## 2. Modeling
- K-fold와 lgbm으로 모델을 구성합니다

```python
# input, output  을 나눈다.

X = train.drop(['date_time', '전력사용량(kWh)'], axis = 1)
y = train['전력사용량(kWh)']
X_test = test.drop('date_time', axis = 1)
```

```python
X.head()
```

```python
y
```

```python
X_test
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
from sklearn.model_selection import StratifiedKFold

# 검증용 20%, 학습용 80% 실행된다.
n_splits = 5
sfk = StratifiedKFold(n_splits = n_splits)

y_pred = np.zeros(len(X_test))

result = 0

# train데이터셋에서 건물의 비율을 동일하게하기 위해 StratifiedKFold 사용
# num 기준으로 k-fold를 한다.
for tr_idx, val_idx in sfk.split(X, X['num']):    
    # 학습용 데이터의 input 값
    X_train = X.iloc[tr_idx, :]
    # 학습용 데이터의 output값
    y_train = y[tr_idx]

    # 검증용 데이터의 input 값
    X_val = X.iloc[val_idx, :]
    # 검증용 데이터의 output 값
    y_val = y[val_idx]

    # n_estimators : 모델의 갯수를 정하는 것
    # learning_rate : 
    # mse : mean squared error (평균제곱근편차)
    lgbm = LGBMRegressor(n_estimators = 10000, learning_rate= 0.1,objective = 'mse')
    
    # early~~ : 모델을 50개씩 늘린다.
    # verbose : 시각적으로 몇개씩 나타내겠다.
    lgbm.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric= 'rmse', early_stopping_rounds= 50,verbose = 400)
    y_pred += lgbm.predict(X_test) / n_splits
    
    # (93215.6 + 49006 + 62012.6 + 58980.1 + 144184) / 5
    # 위의 결과값들의 평균을 계산하는 코드를 아래에 작성해보세요.
    val_pred = lgbm.predict(X_val)
    # 검증값에 예측값을 빼서 result 에 넣어준다.
    # 평균을 구하고, 루트값, 제곱값을 구한다.
    # result += np.mean(np.sqrt(np.square(y_val-val_pred))) / n_splits
    # result += np.mean((y_val-val_pred)**0.5) / n_splits
    
    # smape 식으로 result를 도출해보세요.
    # result += np.mean((np.abs(y_val-val_pred))/(np.abs(y_val) + np.abs(val_pred))) / n_splits
    result += (np.mean(np.abs(y_val-val_pred)/(np.abs(y_val)+np.abs(val_pred))) * 100) / n_splits
    
    print("=" * 20)
    print(result)
    print("=" * 20)
```
