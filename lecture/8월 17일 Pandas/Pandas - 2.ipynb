## 1. 데이터 삽입 및 삭제 (DataFrame, drop)
- 칼럼에 따라 행과 혹은 인덱스에 따라 열로 데이터를 삭제할 수 있따.

```python
user_id = [12, 55, 104, 109, 120, 130, 297, 305]
age = [14, 22 ,18, 25, 39, 41, 18, 56]
name = ["ko","kiwi","ahn","ma","wisly","taylor","oh","yu"]
mail = ["ko@gmail.com","kiwi@gmail.com","ahn@gmail.com","ma@gmail.com",
        "wisly@gmail.com","taylor@gmail.com","oh@gmail.com","yu@gmail.com"]

data = {
    "user_id" : user_id,
    "age" : age,
    "name" : name,
    "mail" : mail
}

data = pd.DataFrame(data)
```

```python
data
```

### 컬럼을 통한 데이터삭제
- axis=1 (열을 축으로 잡는다)

```python
data = data.drop(['age'], axis=1)
```

```python
data
```

### 인덱스를 통한 데이터 삭제
- axis=0 (행을 축으로 잡는다)

```python
data = data.drop([4], axis=0)
```

```python
data
```

## 2. 데이터 병합 (merge)

```python
date = ['07-23','07-24','07-25','07-26','07-28']
ids = [109,104,297,55,12]
price = [2000,3000,1500,10000,2900]
state = ['canceled','confirmed','confirmed','canceled','confirmed']

df1 = {
    "date" : date,
    "ids" : ids,
    "price" : price,
    "state" : state
}

df1 = pd.DataFrame(df1)
```

```python
df1
```

```python
user_id = [12, 55, 104, 109, 120, 130, 297, 305]
age = [14, 22 ,18, 25, 39, 41, 18, 56]
name = ["ko","kiwi","ahn","ma","wisly","taylor","oh","yu"]
mail = ["ko@gmail.com","kiwi@gmail.com","ahn@gmail.com","ma@gmail.com",
        "wisly@gmail.com","taylor@gmail.com","oh@gmail.com","yu@gmail.com"]

df2 = {
    "user_id" : user_id,
    "age" : age,
    "name" : name,
    "mail" : mail
}

df2 = pd.DataFrame(df2)
```

```python
df2
```

```python
DF = pd.merge(df1, df2, left_on = 'ids', right_on = 'user_id', how='inner')
```

```python
DF
```

## Pivot Table , Groupby
- 데이터의 결과를 보고 싶을 때 사용한다.
- describe()도 피벗테이블이라고 볼 수 있따.

```python
tips = sns.load_dataset("tips")
```

```python
tips.describe()
```

```python
pd.pivot_table(tips, index=['smoker'], values=['tip'], aggfunc=['mean', 'sum', 'std'])
```

## pivot table 과 groupby 의 차이
- groupby는 데이터에 직접접근해서 사용한다.
- 대신 func은 하나만 사용 가능하다.
- 속도면에서는 훨씬 빠르다.
- 각각의 데이터를 분활 검색할 때 용이하다.

```python
tips.groupby(['sex','smoker'])['tip', 'total_bill'].mean()
```

## for 문으로 index 만들기
- 조건에 따라 값을 대입시켜서 새로운 index를 생성한다.

```python
 rows = tips.shape[0]
    
for i in range(rows) :
    tips.loc[i, 'index_id'] = str(2 * i)
    
tips
```
