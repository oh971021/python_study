```python
from bs4 import BeautifulSoup 
import urllib.request 
from urllib.request import urlopen
from urllib.parse import quote
import pytagcloud
import re

from konlpy.tag import Okt
from collections import Counter

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
keyword = input(str())
url = 'https://news.joins.com/search/?keyword=' + quote(keyword)
print(url)
```

```python
# url 의 html 을 긁어오는 함수
html = urlopen(url)
html
```

```python
soup = BeautifulSoup(html.read(), 'html5lib', from_encoding = 'utf8')
```

```python
print(soup)
```

#### article뒤에 *은 어떤 문자가 들어오든 받겠다 라는 뜻

```python
# 인덱스는 0~ 25 까지 범위만 작동한다.
i = 1
# url list 를 가지고 온다.
# 즉, 기사가 list 화 되어서 저장이 되어있어서 인덱스 값(i)으로 기사를 찾아올 수 있다.
article_url = soup.findAll("a", {"href":re.compile("https://news.joins.com/article*")})[i]["href"]
article_url
```

```python
# article을 열고, soup2를 생성한다.
html2 = urlopen(article_url)
# soup2 에는 html2에서 열었던 html을 
soup2 = BeautifulSoup(html2.read(), 'html5lib', from_encoding = 'utf8')
print(soup2)
```

```python
# find 는 싹 다 가져오고, select 는 일부만 가져온다.
# contents 는 많은 html 중에 article_body만 가져왔다.
# 현재 div 클래스에 article_body 라는 녀석을 끌어왔다.
contents = soup2.select('div.article_body')
contents
# class 나 id 는 웹사이트 별로 다를 수 있다.
# 그래서 내가 필요한 자료를 가져오려면 확인해봐야한다.
```

```python
raw = ""
for title in soup.findAll("a", {"href":re.compile("https://news.joins.com/article*")}):
    article_url = title['href'] #순서대로  url 주소 가져오기
    
    try : 
        # for 문에서 가져온 url 에 대한 기사를 html2로 끌어온다.
        html2 = urlopen(article_url)#html로 url긇어오는 부분
        soup2 = BeautifulSoup(html2.read(), 'html5lib', from_encoding = 'utf8')
        contents =soup2.select('div.article_body')
        for i in contents: #기사에서 내용을 가져옴
            # i 라는 article 페이지에서
            # Text 형식의 파일을 모두 가지고 오겠다.
            data = str(i.findAll(text=True)) #내용의 텍스트를 문자열로 저장
            raw = raw + data
    except Exception as e :
         pass #오류발생시 pass
print(raw)
```

```python
nlp = Okt()
# nouns 는 raw를 가진 문자열 list가 된다.
nouns = nlp.nouns(raw)
nouns
```

```python
nouns[0]
```

```python
result = []
for i in nouns:
    if len(i) > 1:
        # append = 더하기
        result.append(i)
result
```

```python
# 단어가 얼마나 찍히는지 확인하는 함수
count = Counter(result)
count
```

```python
# 가장 많이 나온 단어 3순위 까지 뽑아내서 tag에 넣는다.
tag = count.most_common(50)
```

```python
# 그림을 그리기 위한 준비단계
# pytag를 사용한다.
# maxsize 는 글자 크기
taglist = pytagcloud.make_tags(tag, maxsize=100)
print(taglist)
```

```python
# 그림을 그린다.
# taglist로 생성한다.
pytagcloud.create_tag_image(taglist, "wordcloud_joins.png", 
                            size=(900, 700), fontname="korean",
                            # 타원형 느낌으로 그린다.
                            rectangular=False)
```

```python
img = mpimg.imread("wordcloud_joins.png")
# 그린 그림을 보여달라
plt.imshow(img)
```

```python
raw2 =  ""

for title in soup3.findAll("p","tit"):
    title_link = title.select('a')
    article_url = title_link[0]['href']
    try : 
        html4 = urlopen(article_url)#html로 url긇어오는 부분
        soup4 = BeautifulSoup(html4.read(), 'html5lib', from_encoding = 'utf8')
        contents =soup4.select('div.article_txt')
        for i in contents: #기사에서 내용을 가져옴
            data = str(i.findAll(text=True)) #내용의 텍스트를 문자열로 저장
            raw2 = raw2 + data
    except Exception as e :
         pass #오류발생시 pass
print(raw2)
```

```python
nlp = Okt()
# nouns 는 raw를 가진 문자열 list가 된다.
nouns2 = nlp.nouns(raw2)
nouns2
```

```python
result2 = []
for j in nouns2:
    if len(j) > 1:
        # append = 더하기
        result2.append(j)
result2
```

```python
# 단어가 얼마나 찍히는지 확인하는 함수
count2 = Counter(result2)
count2
```

```python
# 가장 많이 나온 단어 3순위 까지 뽑아내서 tag에 넣는다.
tag = count2.most_common(50)
```

```python
# 그림을 그리기 위한 준비단계
# pytag를 사용한다.
# maxsize 는 글자 크기
taglist = pytagcloud.make_tags(tag, maxsize=100)
print(taglist)
```

```python
# 그림을 그린다.
# taglist로 생성한다.
pytagcloud.create_tag_image(taglist, "wordcloud_joins.png", 
                            size=(1600, 1000), fontname="korean",
                            # 타원형 느낌으로 그린다.
                            rectangular=False)
```

```python
img = mpimg.imread("wordcloud_joins.png")
# 그린 그림을 보여달라
plt.imshow(img)
```
