# 1.1 데코레이터 함수 정의

### 데코레이터 함수를 정의
- 하나의 매개변수는 함수를 받는다.
- 함수를 그대로 전달한다.

````python
# 데코 함수는 = 함수를 만들어서 함수를 받고, 그대로 함수를 반환하는 것
# 첫번째 받는 함수는 매개변수가 하나여야 한다. (함수를 그대로 반환할 것이기 떄문에)
def func(func_) :
    print("데코한번한다")
    return func_
````

````python
def add(x,y) :
    return x +y
````

````python
# 데코레이팅 : 함수에 함수를 전달해서 함수를 반환하는 것.
f = func(add)
````

````python
f(10,20)
````

### 데코레이터 함수 내에 내부 함수를 정의해서 새로운 기능 추가

- 함수를 전달 받아 지역변수로 처리
- 내부 함수를 반환한다.
- 내부 함수 내에서 저장된 함수를 실행한다. 

````python
# 내부함수를 만들고, 내부함수 안에서 데코레이팅 했던것 처럼 원래의 함수를 그대로 반환한다.
def dec(func):
    print("데코 1번")
    def new_func():
        print('데코 2번')
        return func()
    return new_func
````

````python
def add() :
    return 10
````


### 데코레이터 함수
- 실행함수를 데코함수에 넣는다.
- 데코 함수로 실행 된 결과도 실행함수와 같은 결과가 나온다.
- 데코레이터를 스택으로 쌓아서 사용하는 경우도 있다.
- @데코레이터함수1 
- @데코레이터함수2
- (실행할 함수) = 실행함수의 인자값이지만, 기능이 추가된다.

## 1.2 데코레이터 실행함수 인자 전달 하기
- 실행함수 저장과 실행함수의 인자를 분리한다.
- 내부 함수는 실행함수의 인자를 받아서 저장된 실행함수를 실행한다.

````python
def dec_1(func): # 2. add_1이 저장된다. 
    print("dec call")
    def new_func(*args, **kwargs): # 4. (100,200) 을 받는다.
        print('new_func call')
        return func(*args, **kwargs) # 5. 저장된 값을 반환한다.
    return new_func # 3. add_1(100,200) 의 인자 (100,200)을 반환한다.
````

````python
# 데코레이팅하면 지연실행된다.
@dec_1 # 1. 데코로 add를 넣는다.
def add_1(x,y) : # 6. 여기서 add_1 함수를 실행한다. (지연실행)
    print("add call")
    print (x + y)
````

````python
add_1(100,200)
````

## 1.3  데코레이터 실행순서

- 데코레이터는 여러 개를 실행함수에 연결할 수 있다.
- 데코레이터 순서는 밑에 정의한 것부터 차례대로 구성된다.


````python
# 합성함수를 구성하는 것.
# 데코레이터 함수를 3개 만들어놨다.
def a(func):
    print("a")
    def new_func():
        print('A')
        func()
    return new_func


def b(func):
    print("b")
    def new_func():
        print('B')
        func()
    return new_func


def c(func):
    print("c")
    def new_func():
        print('C')
        func()
    return new_func
````

### 데코레이터 적용 순서 
- 로딩시 자동으로 데코레이터가 구성된다.

````python
@a
@b 
@c # <- 우선 적용
def func():
    pass

# 결과값은 func 변수에 들어있다.
````

### 실행순서 
- 실행슨서는 제일 위에 정의한 데코레이터부터 순차적으로 실행된다.


````python
func()
````

````python
def func_1():
    pass

# 데코레이터 얹는거랑 같은 기능.
func_1 = a(b(c(func_1))) # 데코레이터 문법 적용시키지 않고 동일한 함수를 생성
````

````python
func_1()
````

# 2.  정보상실 문제 해결

- 데코레이터를 구성하면 내부함수가 전달되므로 저장된 실행함수의 정보가 외부에 전달되지 않는다.
- 외부함수의 정보를 전달하려면 함수 내의 메타정보를 갱신해야 한다.

* 덮어쓰기

`__name__`, `__module__`, `__qualname__`, `__doc__`, `__annotations__`

* 업데이트

`__dict__`

* 원본함수 저장

`__wrapped__`

## 2.1  내부 함수가 전달되어 기존 함수에 대한 정보가 없음

````python
def outer(func):
    def inner(*args, **kwargs):
        print("함수가 실행됩니다.")
        return func(*args, **kwargs)
    return inner
````

````python
@outer
def func_3(a, b, c):
    """3개의 숫자 또는 Sequence를 받아서 + 연산 후 리턴"""
    return a + b + c
````

````python
func_3(1, 2, 3)
````

````python
func_3('a', 'b', 'c')
````

## 2.2 wraps 구현
- 실행함수의 정부를 데코레이터를 사용해서 갱신하기


````python
# 함수 관련된 기능을 모아둔 툴 (functools)
from functools import wraps

def dec(func):
    # 이너함수에 씌워줘야한다.
    # 전달된 함수의 정보에 갱신한다.
    # 데코레이터 함수는 바로위에 붙혀준다.
    
    @wraps(func) # 메타 정보 관리
    def wrapper(*args, **kwargs): # *: 가변인자 / **: 키워드인자
        result = func(*args, **kwargs)
        return result
    return wrapper

````

````python
@dec # @dec 사용하지 않을 경우: add = dec(add) ; def add(x, y): return x + y ;
def add(x, y):
    return x + y
````

````python
@dec
def multiply(x, y):
    return x * y
````

````python
print(add(5, 5))
````

````python
add.__name__
````

````python
print(multiply(4, 3))
````

````python
multiply.__name__
````

## 2-3  더 깊은 데코레이터

- 데코레이터를 처리하는 경우 데코레이터에도 별도의 매개변수를 받을 수 있다.
- 이때 제일 밖의 함수는 데코레이터의 매개변수를 받고 그 내부 함수부터 함수를 받도록 처리한다.

````python
def make_deco(value):
    def outer(func):                    # 여기서부터
        def inner(*args, **kwargs):
            print(value)
            return func(*args, **kwargs)
        return inner                     # 여기까지가 데코레이터
    return outer
````

````python
# 데코레이터 함수에서 사용할 매개변수를 넣어줄 수 있다.
@make_deco('!!!@#!@#!@#!@#$@#$%@!#$%@#$')
def func():
    print("func!!!")
````

````python
func()
````

### 최상위 테코레이터 함수에 인자를 받아서 처리합니다.

````python
def typecheck(*t_args):# t_args
    def outer(func):
        def inner(*args, **kwargs):
            for i, t_arg in enumerate(t_args):
                if t_arg != type(args[i]):
                    raise TypeError("정확하지 않은 타입입니다.")
            return func(*args, **kwargs)
        return inner
    return outer
````

````python
@typecheck(int, int, int)
def sum(a, b, c):
    return a + b + c
````

````python
sum(1, 2, 3)
````

## 타입체크 예제 ( 데코레이터 함수 )

````python
# l 누르면 라인이 생긴다.
````

````python
def typecheck1(*tt_args):
    def outer1(func):
        def inner1(*args, **kwargs) :
            for i, tt_arg in enumerate(tt_args): # 튜플로 돌아간다.
                if tt_arg != type(args[i]): 
                    # 타입 에러의 메세지를 정해놓을 수 있다.
                    raise TypeError("정확치 않은 타입입니다")
            return func(*args, **kwargs)
        return inner1
    return outer1
````

````python
@typecheck1(int, int, int) # 타입확인용 타입을 넣어놓는다.
def sum(a, b, c):
    return a + b + c
````

````python
# 이 친구들의 타입을 체크한다.
sum(1,2,3)
````

### 최상위 테코레이터 함수에 인자를 받아서 처리합니다.

````python
## 키워드인자들도 반응하도록 
# 모든 인자를 다 받아도 체크가 가능한 로직.

def typecheck_1(*t_args, **t_kwargs):# t_args
    # 가변 인자로 파라미터를 받고, 몇개가 들어오든 타입체크를 해준다.
    
    def outer(func):
        def inner(*args, **kwargs):
            # 위치인자 검증
            for i, t_arg in enumerate(t_args):
                if t_arg != type(args[i]):
                    raise TypeError("정확하지 않은 타입입니다.")
                    
            # 키워드인자 검증
            for key in t_kwargs.keys():
                if t_kwargs[key] != kwargs[key].__class__:
                    raise TypeError("정확하지 않은 타입입니다.")
            
            return func(*args, **kwargs)
        return inner
    return outer
````

````python
@typecheck_1(a=int, b=int, c=int)
def sum(a, b, c):
    return a + b + c
````

````python
sum(a=3, b=2, c=3)
````
