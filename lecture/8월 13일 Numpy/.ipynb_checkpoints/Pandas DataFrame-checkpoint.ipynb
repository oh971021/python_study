### Apply 함수
- 해당 함수는 데이터 프레임에서 특정 칼럼의 값을 조건에 맞춰 바꾸는 기능
- 즉, 내가 원하는대로 조건식을 만들어놓고, 해당 칼럼의 모든 값에 적용해서 바꾼다.

```python
def in_business_age_name(Age):
    name=''
    if Age <= 25 : name = '사회초년생'
    elif Age <= 35 : name = '사회생활 1~10년차'
    elif Age <= 45 : name = '관리자 역할'
    else : name = '은퇴예정자'

    return name
```

- 위의 함수는 파이썬 문법인 '함수 정의'이고, 이 함수를 칼럼 변경 기준으로 사용한다.

```python
df['나이대'] = df['나이'].apply(lambda x : in_business_age_name(x))
```

#### 칼럼.apply(lambda 결과값: 함수이름(인자값))

- 해당하는 조건에 맞춰 함수를 통해 조건을 건다.
- 람다 함수의 x 에는 칼럼의 value 가 들어간다.

```python
df
```

## DataFrame 합치기

### Merge

```python
df_first = pd.DataFrame({'id':[10,11,12,13,14],
                                'python':[1,2,3,4,5],
                                'C':[2,3,4,5,6],
                                'C++':[2,3,4,1,2]})
                                
df_second = pd.DataFrame({'id':[10,11,12,13,14],
                                 'java':[5,2,3,3,1],
                                 'javascript':[1,3,2,1,4],
                                 'Php':[1,3,2,0,0]})
```

```python
df_first
```

```python
 df_second
```

```python
# 방법 1
df_first.merge(df_second)
```

```python
#방법 2
pd.merge(df_first,df_second)
```

### Join

```python
# id 값 중 뒷 부분이 기존의 데이터프레임과 다르다.
df_third = pd.DataFrame({'id':[10,11,12,17,18],
                        'html':[1,2,3,4,5],
                        'css':[5,4,3,2,1],
                        'go':[1,3,2,4,2]})
```

```python
pd.merge(df_third,df_second)
```

- 현재 칼럼의 value 중 id의 값인 17, 18 은 겹치지 않는 값이라서 표시되지않는다.

### How

```python
pd.merge(df_second, df_third, how='outer')
```

- 두 데이터프레임을 outer 방법으로 merge를 시키면 모든 데이터프레임의 값이 나온다.
- second는 id != 17,18 / third는 id != 13,14 였음에도 불구하고, 모든 id 값을 써주면서 각각의 없는 값들은 NaN을 표시해준다.

### Merge - on

```python
data1 = pd.DataFrame({'lkey':['foo','bar','baz','foo'],
                    'value':[1,2,3,5]})

data2 = pd.DataFrame({'rkey':['foo','bar','baz','foo'],
                     'value':[5,6,7,8]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,left_on='lkey',right_on='rkey')
```

- lkey와 rkey를 기준으로 데이터프레임을 병합했다.
- 두 l, r key는 같은 값이므로, 나중에 데이터 전처리 작업 시 둘 중 하나를 제거하는 방식으로 활용하면 된다.
- key 값으로 병합했기 때문에 value는 자동으로 x,y 로 나뉘게 된다.

- x, y가 맘에 안들면 merge() 옵션 중 suffixes 라는 옵션을 통해 직접 이름을 할당할 수 있는데 이건 다음에 알아보자.
