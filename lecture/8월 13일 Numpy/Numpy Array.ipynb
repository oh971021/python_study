```python
import numpy as np
```

## Array 합치기

```python
a = np.array([[1,2],[3,4]])
```

```python
b = np.array([[5,6]])
```

```python
a, b
```

### aixs 란??

```python
np.concatenate((a,b))
```

```python
np.concatenate((a,b), axis=0)
```

- 두 행렬을 axis=0 기준으로 합친다면 a array의 아랫 부분에 가로로 합쳐진다.

```python
np.concatenate((a,b), axis=1)
```

- 에러 내용 : input 차원은 정확히 axis에 대해 매칭 되어야 한다.
- 즉, 현재 어느 칼럼으로 붙을지 확실치가 않다는 것

```python
b.T
```

### 행렬 전치
- Transpose or T 를 사용한다.

```python
np.concatenate((a,b.T),axis=1)
```
