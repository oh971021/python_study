## Pandas를 활용한 데이터 전처리

```python
tips1.loc[0,'time'] = 'Lunch'
tips1.head()
```

```python
tips.head()
```

```python
tips.loc[tips['tip']>5,'level'] = 'good'
tips.loc[tips['tip']<=5,'level'] = 'bad'
tips
```

```python
tips.loc[tips['level']!='good']
```

```python
rows = tips.shape[0]
print(rows)
for i in range(rows):
    tips.loc[i,'index_id'] =str(2 * i) # str 문자열로 변환
tips    
```

```python
tips_m = tips.set_index('index_id')
tips_m.head()
```

```python
tips_m.loc['0']
```

```python
tips_m.iloc[0]
```

```python
tips.head()
```

```python
tips.iloc[:1]
```

```python
tips.loc[:1]
```

```python
tips['ratio'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
def bill(x):
    total_bill = x['total_bill']
    tip = x['tip']
    bill = total_bill - tip
    return bill

tips['bill'] = tips.apply(bill, axis = 'columns')
tips
```

```python
# 1문제 tips.loc[tips['ratio']>=0.2]
```

```python
# one-hot encoding
tips.loc[tips['level']=='good','good'] = 1
tips.loc[tips['level']=='bad','bad'] = 1
tips
```

```python
tips[tips['good'].isnull()]
```

```python
tips['good'] = tips['good'].fillna(0)
tips['bad'] = tips['bad'].fillna(0)
tips
```

```python
tips.loc[(tips['total_bill']>24)&(tips['tip']>3.5),'customer'] = 'vip'
tips.loc[(tips['total_bill']<=24)|(tips['tip']<=3.5),'customer'] = 'non-vip'
tips
```

```python
tips['total_bill(pre)'] = tips['total_bill'].shift(+1)
tips['total_bill(next)'] = tips['total_bill'].shift(-1)
tips
```

```python
a = ['index_id',
     'total_bill(pre)',
     'total_bill(next)'
    ]
tips_simple = tips.drop(a, axis = 'columns')
tips_simple
```

## pivot table과 groupby

```python
index = ['level', 'smoker']
values = ['tip', 'bill']
aggfunc = ['mean','sum','std']
pd.pivot_table(tips, index = index ,values = values, aggfunc = aggfunc)
```

```python
# pivot table과 group by가 상호 호환적임
# group by가 pivot table 보다 10배 정도 빠르며 group by는 추가 연산,데이터베이스 저장,파일로 저장,머신러닝 알고리즘에 적용
# pivot table은 최종적으로 결과 도출해서 보여줄 때만 사용
index = ['level', 'smoker']
tips.groupby(index)['tip', 'bill'].mean()
```
